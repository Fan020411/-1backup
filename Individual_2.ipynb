```python
# Create a label encoder instance and encode the 'article' column
label_encoder = LabelEncoder()
daily_sales_article['article_Code'] = label_encoder.fit_transform(daily_sales_article['article'])

# Create a dictionary to map the encoded numbers to their corresponding countries
article_code_mapping = dict(zip(label_encoder.classes_, range(len(label_encoder.classes_))))

# Print the mapping
print("article Code Mapping:")
print(article_code_mapping)
```

These steps ensure that the data is properly formatted and prepared for training and testing with the DeepAR algorithm, allowing for seamless integration and analysis within the SageMaker environment.

```python
def convert_to_deepar_json(df, filename, is_test=False):
    grouped = df.groupby('article_Code')
    time_series = []
    for _, group in grouped:
        # Create a record for each time series
        record = {
            "start": str(group['date'].min()),
            "target": group['Sales'].tolist(),
            "cat": [int(group['article_Code'].iloc[0])]
        }
        if is_test:
            # For the test set, include only the last 'prediction_length' points
            record["target"] = record["target"][-prediction_length:]
        time_series.append(record)
    
    # Write the time series to a JSON Lines file
    with open(filename, 'w') as f:
        for ts in time_series:
            f.write(json.dumps(ts) + '\n')

# Set the prediction length
prediction_length = 7  # Adjust this value as needed

# Split the data into training and test sets
train_df = daily_sales_article.copy()
test_df = daily_sales_article.groupby('article_Code').apply(lambda x: x.tail(prediction_length)).reset_index(drop=True)

# Ensure that the test set has the last 'prediction_length' points for each time series
test_df = test_df.groupby('article_Code').filter(lambda x: len(x) == prediction_length)

# Convert the training and test sets to DeepAR JSON format
convert_to_deepar_json(train_df, 'train_data.json')
convert_to_deepar_json(test_df, 'test_data.json', is_test=True)
```

```python
# Create a SageMaker session for managing interactions with the SageMaker environment
sagemaker_session = sagemaker.Session()

# Retrieve the execution role for the current SageMaker notebook instance
role = get_execution_role()

# Retrieve the AWS region name from the SageMaker session
region = sagemaker_session.boto_region_name

# Specify the DeepAR container
image_uri = retrieve('forecasting-deepar', region)

# Define the default Amazon S3 bucket for the SageMaker session
bucket = sagemaker_session.default_bucket()

# Define the prefix for the S3 key where data will be stored
prefix = 'sagemaker/bakerysales/category-prediction'

# Define the path to the local training data file
local_train = 'train_data.json'

# Define the path to the local test data file
local_test = 'test_data.json'

# Upload the local training data file to the specified S3 bucket and prefix, returning the S3 URI
train_uri = sagemaker_session.upload_data(local_train, bucket=bucket, key_prefix=prefix)

# Upload the local test data file to the specified S3 bucket and prefix, returning the S3 URI
test_uri = sagemaker_session.upload_data(local_test, bucket=bucket, key_prefix=prefix)

# Print the execution role ARN to the console
print("Execution Role ARN:", role)

# Print the AWS region name to the console
print("AWS Region:", region)

# Print the S3 URI for the training data to the console
print("Training Data S3 URI:", train_uri)

# Print the S3 URI for the test data to the console
print("test Data S3 URI:", test_uri)
```

# Estimator Configuration for DeepAR Algorithm Training

This section is about setting the hyperparameters that outlines the configuration process for leveraging Amazon SageMaker's DeepAR algorithm to address CRAVE Coffee House & Bakery's sales forecasting challenges. It defines key parameters tailored to CRAVE's business-side needs, such as data frequency, context length (here i changed the default index from 30 to 150 days to eliminate the adversely impact on the training job), and epoch, ensuring accurate predictions aligned with the business's dynamics (here i altered the defualt setting from 20 to 1o in order to minimize the RMSE). By setting hyperparameters and enabling automatic feature detection, the code aims to optimize model performance and adaptability. Overall, it signifies CRAVE's strategic move towards data-driven decision-making to enhance efficiency and customer satisfaction.

```python
# Define an Estimator for the DeepAR algorithm in SageMaker
deepar_estimator = Estimator(
    sagemaker_session=sagemaker_session,  # Specify the SageMaker session
    image_uri=image_uri,  # Specify the container image URI for the DeepAR algorithm
    role=role,  # Specify the IAM role ARN used by SageMaker
    instance_count=1,  # Specify the number of instances to use for training
    instance_type='ml.m5.large',  # Specify the instance type for training
    output_path=f's3://{bucket}/{prefix}/output',  # Specify the S3 location for saving the model artifacts and output
    base_job_name='bakerysales-deepar-forecasting'  # Specify the base job name for the training job
)

# Set hyperparameters for the DeepAR algorithm
deepar_estimator.set_hyperparameters(
    time_freq='D',  # Specify the time frequency of the data (e.g., 'D' for daily)
    context_length= 150,  # Specify the length of the context used for training (e.g., 30 days)
    prediction_length=7,  # Specify the length of the prediction horizon (forecasting 7 days into the future)
    epochs=10,  # Specify the number of training epochs
    early_stopping_patience=10,  # Specify the number of epochs with no improvement before early stopping
    num_dynamic_feat='auto',  # Automatically detect dynamic features (if any)
    cardinality='auto'  # Automatically detect the cardinality of categorical features (if any)
)
```

# Model Training with DeepAR Estimator

This section establishes data channels for training the DeepAR estimator, facilitating the ingestion of training and test data stored in S3 bucket. By mapping the 'train' and 'test' keys to the respective S3 URIs, it organizes the data for model training and evaluation. Then, the DeepAR estimator is fitted using the specified data channels, enabling the model to learn from the training data and validate its performance against the test data. This process enhances the accuracy and robustness of the DeepAR forecasting model, enabling CRAVE to make informed decisions based on reliable predictions.

```python
# Define data channels for training the DeepAR estimator
# Mapping 'train' and 'test' keys to the S3 URIs of the training and test data
data_channels = {
    'train': train_uri,  # Specify the S3 URI for the training data
    'test': test_uri     # Specify the S3 URI for the test data
}

# Fit the DeepAR estimator using the specified data channels
# This step initiates the training job in SageMaker
deepar_estimator.fit(inputs=data_channels)
```

The below coding initializes and configures a SageMaker endpoint for deploying a trained model. It first specifies the endpoint name and configuration name, then checks if they already exist. By following this, it deploys the trained model to the endpoint with specified instance configurations, serializers, and deserializers. # Preparing Prediction Input for DeepAR Model

## Make Inference: 
If CRAVE utilize the DeepAR model deployed on AWS SageMaker, the shop is able to forecast daily sales volume for specific bakery items within specific time period with remarkable precision. For instance, when predicting sales volume for baguettes (article code 4), the model provides us with a forecast sales volumes for the upcoming 30 days, such as 27.90 units on August 31st. These predictions are derived from historical sales data by capturing exteral factors like past sales volumes, promotional activities, and seasonal trends. Leverage this predictive model, CRAVE will gain valuable insights into their future demand patterns, which empowers them to optimize inventory management, adjust staffing schedules, and allocate budgets effectively. Overall, the time-series predictive model serves as a robust and strategic tool for CRAVE, enables a data-driven decision-making and enhances operational efficiency. 

Moreover, the purpose of of prediction on an 80% interval alongside the mean sales forecast is to provide a range of likely outputs, which allows stakeholders to understand the uncertainty associated with the prediction outcomes and make informed decisions with regard on marketing strategies and daily operations. 





```python
# Define the endpoint name
endpoint_name = 'bakery-sales-endpoint-cat'

# Display the article code list
print("article Code List:")
for article, code in article_code_mapping.items():
    print(f"{article}: {code}")

# Prompt for the country code
selected_article_code = int(input("Enter the article code for prediction: "))

# Find the country name corresponding to the selected country code
selected_article_name = [article for article, code in article_code_mapping.items() if code == selected_article_code][0]

# Prepare the input data for prediction
# 'daily_sales_article' DataFrame contains the historical sales data for all articles
selected_article_sales = daily_sales_article[daily_sales_article['article_Code'] == selected_article_code]  # Filter for selected article

selected_article_sales.tail()

last_date = selected_article_sales['date'].max().strftime('%Y-%m-%d')  # Get the last date of sales for the selected country
prediction_input = {
    "instances": [{"start": "2022-8-31", "target": selected_article_sales['Sales'].tolist(), "cat": [selected_article_code]}],
    "configuration": {"num_samples": 100, "output_types": ["mean", "quantiles"], "quantiles": ["0.1", "0.9"]}
}

# Invoke the endpoint for prediction
sagemaker_runtime = boto3.client('sagemaker-runtime')
response = sagemaker_runtime.invoke_endpoint(
    EndpointName=endpoint_name,
    ContentType='application/json',
    Body=json.dumps(prediction_input)
)

# Parse the prediction response
predictions = json.loads(response['Body'].read().decode())
predicted_means = predictions['predictions'][0]['mean']
predicted_quantiles = predictions['predictions'][0]['quantiles']
predicted_lower_quantile = predicted_quantiles['0.1']
predicted_upper_quantile = predicted_quantiles['0.9']

# Create a DataFrame for the predicted sales
predicted_dates = pd.date_range(start='2022-08-31', periods=len(predicted_means), freq='D')
predicted_sales = pd.DataFrame({'Date': predicted_dates, 'Predicted Sales': predicted_means})

# Display the predicted sales
print(predicted_sales)

# Filter the historical data to include only the last 30 days for the selected country
last_30_days_sales = selected_article_sales[selected_article_sales['date'] >= (selected_article_sales['date'].max() - pd.Timedelta(days=30))]

# Plot the daily sales for the last 30 days for the selected country
plt.figure(figsize=(16, 6))
plt.plot(last_30_days_sales['date'], last_30_days_sales['Sales'], marker='o', linestyle='-', label=f'Historical Sales ({selected_article_name})')

# Plot the predicted mean sales
plt.plot(predicted_dates, predicted_means, label='Predicted Mean Sales')

# Plot the predicted quantiles as a shaded area
plt.fill_between(predicted_dates, predicted_lower_quantile, predicted_upper_quantile, color='gray', alpha=0.5, label='Predicted 80% Confidence Interval')

# Add labels and title
plt.title(f'Daily Sales with Prediction for December 1st Onwards ({selected_article_name})')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

# Delete the Endpoint and Endpoint Config

This workflow ensures that the deletion of endpoints and configurations is intentional and prevents accidental deletion with the function of maunully enter whether delete the endpoint or not. The automation of deleting process and promoting for confirmation largely streamlines endpoint management for CRAVE, and therefore reduces the risk of errors and ensures intentional actions. This methodology enhances operational efficiency and minimizes disruptions, ultimately optimizes CRAVE's analytics infrastrcture. 



```python
# Prompt for confirmation before deleting the endpoint and endpoint configuration
confirmation = input("Are you sure you want to delete the endpoint and endpoint configuration? (yes/no): ")

if confirmation.lower() == 'yes':
    # Check if the endpoint configuration already exists
    try:
        sagemaker_client.describe_endpoint_config(EndpointConfigName=endpoint_config_name)
        print(f'Endpoint configuration "{endpoint_config_name}" already exists. Deleting...')
        sagemaker_client.delete_endpoint_config(EndpointConfigName=endpoint_config_name)
        print(f'Endpoint configuration "{endpoint_config_name}" deleted.')
    except ClientError as e:
        if e.response['Error']['Code'] == 'ValidationException':
            print(f'Endpoint configuration "{endpoint_config_name}" does not exist.')
        else:
            raise

    # Check if the endpoint already exists
    try:
        sagemaker_client.describe_endpoint(EndpointName=endpoint_name)
        print(f'Endpoint "{endpoint_name}" already exists. Deleting...')
        sagemaker_client.delete_endpoint(EndpointName=endpoint_name)
        print(f'Endpoint "{endpoint_name}" deleted.')
    except ClientError as e:
        if e.response['Error']['Code'] == 'ValidationException':
            print(f'Endpoint "{endpoint_name}" does not exist.')
        else:
            raise
else:
    print("Deletion cancelled.")
```
