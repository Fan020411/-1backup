The purpose of this section is to establish a secure and correctly configured environment for developing and deploying machine learning models with Amazon SageMaker.
```python
# Create a SageMaker session, which manages interactions with SageMaker services
sagemaker_session = sagemaker.Session()

# Get the IAM execution role used for SageMaker to access AWS resources
role = get_execution_role()

# Get the AWS region associated with the SageMaker session
region = sagemaker_session.boto_region_name
```

# Prepare and Inspect Data

The method of getting ready and checking the project's input advertising data is shown in this part. The process involves importing the dataset, addressing any missing values, eliminating superfluous columns, and arranging the data into the x and y variables. 
To ensure that the model can be trained and assessed on distinct segments of the data, hence improving its capacity to generalize to new, unseen data, it is imperative to partition the dataset into training and validation sets. The training and validation datasets are then combined again into distinct DataFrames, features are aligned with their corresponding labels, and the files are saved as CSV files without headers or indices in order to conform to the format required by AWS SageMaker. After uploading these files to an Amazon S3 bucket, establishing the groundwork for training model on SageMaker. This preparation stage is indispensable, ensuring that the data is clean, well-structured, and stored in a manner that facilitates efficient model training and validation for the following procedures. 

```python
# Import the pandas library as 'pd' for data manipulation and analysis
import pandas as pd

# Import the train_test_split function from scikit-learn for data splitting
from sklearn.model_selection import train_test_split
```

```python
advertising_data = pd.read_csv('Advertising.csv')
```

```python
advertising_data.isna()
```

```python
advertising_data = advertising_data.dropna()
advertising_data = advertising_data.drop('Unnamed: 0', axis=1)
```

```python
# The target variable column is named 'TV'

# Create the feature matrix 'X' by dropping the 'TV' column
X = advertising_data.drop('Sales', axis=1)

# Create the target variable 'y' by converting 'TV' to integers
y = advertising_data['Sales'].astype('float')

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Concatenate the features and labels back into one DataFrame for training data
advertising_train_data = pd.concat([y_train, X_train], axis=1)

# Concatenate the features and labels back into one DataFrame for validation data
advertising_validation_data = pd.concat([y_val, X_val], axis=1)

# Save the training data to a CSV file without headers and indices
advertising_train_data.to_csv('advertising_train.csv', header=False, index=False)

# Save the validation data to a CSV file without headers and indices
advertising_validation_data.to_csv('advertising_validation.csv', header=False, index=False)
```

```python
# Define your Amazon S3 bucket and prefix for data storage
bucket = sagemaker_session.default_bucket()
prefix = 'sagemaker/advertising/classification'

# Paths to your local data files - replace with your actual file paths
local_train = 'advertising_train.csv'
local_validation = 'advertising_validation.csv'

# Upload the local training data to the specified S3 bucket and prefix
train_uri = sagemaker_session.upload_data(local_train, bucket=bucket, key_prefix=prefix)

# Upload the local validation data to the specified S3 bucket and prefix
validation_uri = sagemaker_session.upload_data(local_validation, bucket=bucket, key_prefix=prefix)
```

```python
# Print the S3 URI for the training data
print("Training URI: ", train_uri)

# Print the S3 URI for the validation data
print("Validation URI: ", validation_uri)
```

# Get the Linear Learner Image URI

This step aims to seamlessly integrate with AWS Sagemaker's envrionment by retrieving the container image URI for the Linear Learner algorithm, which is specific to the AWS region where my SageMaker session is configured to operate. And facilitate a streamlined, regino-specific deployment of ML models, ensures compatibility and performance efficiency within the environment.  



```python
from sagemaker import image_uris

# Retrieve the container image URI for the SageMaker Linear Learner algorithm
container = image_uris.retrieve(framework='linear-learner', region=region)
```

# Configure the SageMaker Linear Learner Estimator

This section get the model prepared by calculating the dataset's size and configuring a SageMaker Linear Learner for regression analysis. It quantifies the data's scale, sets up the estimator with specifics like instance type and hyperparameters, including the number of input features and batch size, aiming to predict advertising outcomes efficiently. This crucial step set stage for integrating data insights with AWS resources for optimized model training.

```python
# Calculate the number of rows and features in the 'Advertising_data' DataFrame
num_rows, num_features = advertising_data.shape

# Print the number of rows and features
print("Number of Rows:", num_rows)
print("Number of Features:", num_features)
```

```python
# Create a SageMaker Linear Learner estimator
linear_learner = sagemaker.estimator.Estimator(container,
                                               role, 
                                               instance_count=1, 
                                               instance_type='ml.m5.large',
                                               output_path=f's3://{bucket}/{prefix}/output',
                                               sagemaker_session=sagemaker_session)

# Set hyperparameters for the Linear Learner
linear_learner.set_hyperparameters(feature_dim=3,  # Number of input features (excluding target)
                                   mini_batch_size=32,  # Size of mini-batches for training
                                   predictor_type='regressor',  # Specify 'regressor' for regression
                                   normalize_data=True,  # Normalize input features
                                   normalize_label=True)  # Normalize target variable for regression
```

# Train the Model

This part initiates a training job in AWS SageMaker with the Linear Learner algorithm, primarily focus on training the model applying the meticulous prepared and pre-pocessed training and validation data. It reflects the previous application of setting hyperparameters and the prepared datasets to optimize model's performance in order to ensure its learns precisely from the provided-datase.



```python
# Fit the SageMaker Linear Learner estimator to the training and validation data
linear_learner.fit({'train': TrainingInput(train_uri, content_type='text/csv'),
                    'validation': TrainingInput(validation_uri, content_type='text/csv')})
```

# Deploy the Model

#### Check to see if the Endpoint exists and if it does delete it
This section demonstrates managing AWS SageMaker endpoints for deploying and undeploying ML Linear Learner models. This deployment process comprises specifying the instance type and count along with a custom endpoint name, which makes the model available for inference effectively. This endpoints management underscores the machine learning model lifecycle in a cloud environment, from deployment for Presotea application to removal when no longer needed, which ensures resource utilization and operational cost control. 


```python
EndpointConfig="regression-linear-learner-endpoint"
Endpoint="regression-linear-learner-endpoint"
```

```python
import boto3

def delete_sagemaker_endpoint(endpoint_name):
    # Initialize SageMaker client
    sagemaker = boto3.client('sagemaker', region_name=region)
    
    try:
        # Check if the endpoint configuration exists
        response = sagemaker.describe_endpoint_config(EndpointConfigName=endpoint_name)
        
        # If the configuration exists, delete it
        if response:
            sagemaker.delete_endpoint_config(EndpointConfigName=endpoint_name)
            print(f"Endpoint configuration '{endpoint_name}' has been deleted.")
        
        # Check if the endpoint exists
        response = sagemaker.describe_endpoint(EndpointName=endpoint_name)
        
        # If the endpoint exists, delete it
        if response:
            sagemaker.delete_endpoint(EndpointName=endpoint_name)
            print(f"Endpoint '{endpoint_name}' has been deleted.")
        
        return True  # Deletion successful
    except Exception as e:
        error_message = str(e)
        if "Could not find endpoint configuration" in error_message:
            print(f"Endpoint configuration '{endpoint_name}' not found. No action taken.")
            return True  # Configuration not found, exit gracefully
        elif "Could not find endpoint" in error_message:
            print(f"Endpoint '{endpoint_name}' not found. No action taken.")
            return True  # Endpoint not found, exit gracefully
        else:
            print(f"Error deleting SageMaker endpoint and configuration: {error_message}")
            return False  # Deletion failed
```

```python
# Delete the Endpoint and Config

result = delete_sagemaker_endpoint(Endpoint)
if result:
    print(f"Endpoint '{Endpoint}' and its configuration have been deleted.")
else:
    print(f"Failed to delete endpoint '{Endpoint}' and its configuration.")
```

```python
import boto3

# Create a SageMaker client to interact with the SageMaker service
sagemaker_client = boto3.client('sagemaker')

# Deploy the Linear Learner model to the SageMaker endpoint
linear_predictor = linear_learner.deploy(
    initial_instance_count=1,  # Number of initial instances
    instance_type='ml.m5.large',  # Type of instance for serving
    endpoint_name=Endpoint  # Custom endpoint name
)
```

# Make Inferences

This section demonstrates managing AWS SageMaker endpoints for deploying and undeploying ML Linear Learner models. This deployment process comprises specifying the instance type and count along with a custom endpoint name, which makes the model available for inference effectively. This endpoints management underscores the machine learning model lifecycle in a cloud environment, from deployment for Presotea application to removal when no longer needed, which ensures resource utilization and operational cost control.

```python
import boto3

def delete_sagemaker_endpoint(endpoint_name):
    # Initialize SageMaker client
    sagemaker = boto3.client('sagemaker', region_name=region)
    
    try:
        # Check if the endpoint configuration exists
        response = sagemaker.describe_endpoint_config(EndpointConfigName=endpoint_name)
        
        # If the configuration exists, delete it
        if response:
            sagemaker.delete_endpoint_config(EndpointConfigName=endpoint_name)
            print(f"Endpoint configuration '{endpoint_name}' has been deleted.")
        
        # Check if the endpoint exists
        response = sagemaker.describe_endpoint(EndpointName=endpoint_name)
        
        # If the endpoint exists, delete it
        if response:
            sagemaker.delete_endpoint(EndpointName=endpoint_name)
            print(f"Endpoint '{endpoint_name}' has been deleted.")
        
        return True  # Deletion successful
    except Exception as e:
        error_message = str(e)
        if "Could not find endpoint configuration" in error_message:
            print(f"Endpoint configuration '{endpoint_name}' not found. No action taken.")
            return True  # Configuration not found, exit gracefully
        elif "Could not find endpoint" in error_message:
            print(f"Endpoint '{endpoint_name}' not found. No action taken.")
            return True  # Endpoint not found, exit gracefully
        else:
            print(f"Error deleting SageMaker endpoint and configuration: {error_message}")
            return False  # Deletion failed
```

```python
# Delete the Endpoint and Config

result = delete_sagemaker_endpoint(Endpoint)
if result:
    print(f"Endpoint '{Endpoint}' and its configuration have been deleted.")
else:
    print(f"Failed to delete endpoint '{Endpoint}' and its configuration.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# 1. Import your dataset
# Assuming your dataset is in a CSV file named 'data.csv'
df = pd.read_csv('Advertising.csv')

# Assuming your dataset has two columns 'x' and 'y'
x = df['TV']
y = df['Sales']

# 2. Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# 3. Plot the scatter plot and regression line
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', marker='o', label='Data Points')

# Plot the regression line
plt.plot(x, slope*x + intercept, color='red', label='Regression Line')

# Function of the line
line_eq = f'y = {slope:.2f}x + {intercept:.2f}'  # Equation of the line
plt.text(max(x) * 0.7, max(y) * 0.9, line_eq, fontsize=12, color='red')

plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('Scatter Plot- TV Vs. Sales')

plt.grid(True)
plt.legend()
plt.show()

```
